```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import seaborn as sns
```

```python
np.random.seed(123)
random.seed(123)
```

```python
num_transactions = 1000
start_date = datetime(2025, 4, 1)
end_date = datetime(2025, 4, 30)
date_range = [start_date + timedelta(days=x) for x in range((end_date - start_date).days + 1)]
```

```python
products = {
    'Smartphone': {'category': 'Electronics', 'unit_price': 599.99},
    'Laptop': {'category': 'Electronics', 'unit_price': 999.99},
    'Headphones': {'category': 'Electronics', 'unit_price': 89.99},
    'T-Shirt': {'category': 'Clothing', 'unit_price': 19.99},
    'Jeans': {'category': 'Clothing', 'unit_price': 49.99},
    'Sneakers': {'category': 'Clothing', 'unit_price': 79.99},
    'Watch': {'category': 'Accessories', 'unit_price': 149.99},
    'Backpack': {'category': 'Accessories', 'unit_price': 39.99}
}


locations = ['New York', 'Los Angeles', 'Chicago', 'Houston']

# Generate synthetic data
data = {
    'TransactionID': range(1, num_transactions + 1),
    'Date': [random.choice(date_range) for _ in range(num_transactions)],
    'Product': [random.choice(list(products.keys())) for _ in range(num_transactions)],
    'Quantity': [random.randint(1, 5) for _ in range(num_transactions)],
    'Location': [random.choice(locations) for _ in range(num_transactions)]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Add Category and UnitPrice
df['Category'] = df['Product'].apply(lambda x: products[x]['category'])
df['UnitPrice'] = df['Product'].apply(lambda x: products[x]['unit_price'])
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
# Sort by Date and TransactionID
df = df.sort_values(['Date', 'TransactionID']).reset_index(drop=True)
```

```python
# Save to CSV
csv_path = 'techtrend_retail_sales_april_2025_new.csv'
df.to_csv(csv_path, index=False)
print(f"CSV file saved as: {csv_path}")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Save to CSV
csv_path = 'techtrend_retail_sales_april_2025.csv'
df.to_csv(csv_path, index=False)
print(f"CSV file saved as: {csv_path}")
df.head()
```

```python
# Load the CSV file
df = pd.read_csv('techtrend_retail_sales_april_2025.csv')
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Display data info
print("Data Info:")
print(df.info())

```

```python
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

```python
# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Total Sales Revenue
total_revenue = df['TotalPrice'].sum()
print(f"Total Revenue: ${total_revenue:.2f}")
```

```python
# Revenue by Category
revenue_by_category = df.groupby('Category')['TotalPrice'].sum().sort_values(ascending=False)
print("Revenue by Category:")
print(revenue_by_category)
```

```python
# Revenue by Product
revenue_by_product = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False)
print("Revenue by Product:")
print(revenue_by_product)
```

```python
# Revenue by Location
revenue_by_location = df.groupby('Location')['TotalPrice'].sum().sort_values(ascending=False)
print("Revenue by Location:")
print(revenue_by_location)
```

```python
# Daily Sales Trend
daily_sales = df.groupby(df['Date'].dt.date)['TotalPrice'].sum()
plt.figure(figsize=(10, 5))
daily_sales.plot(kind='line', title='Daily Sales Revenue (April 2025)', ylabel='Revenue ($)', xlabel='Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Quantity Sold by Product
quantity_by_product = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
quantity_by_product.plot(kind='bar', title='Total Quantity Sold by Product', ylabel='Quantity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sales by Category (Pie Chart)
category_sales = df.groupby('Category')['TotalPrice'].sum()
plt.figure(figsize=(7, 7))
category_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales Distribution by Category')
plt.ylabel('')
plt.show()
```

```python
# Sales by Location
location_sales = df.groupby('Location')['TotalPrice'].sum()
plt.figure(figsize=(8, 5))
location_sales.plot(kind='bar', title='Sales by Location', ylabel='Revenue ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df['DayOfWeek'] = df['Date'].dt.day_name()
day_sales = df.groupby('DayOfWeek')['TotalPrice'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.figure(figsize=(8, 5))
day_sales.plot(kind='bar', title='Sales by Day of Week', ylabel='Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##UNIVARIATE ANALYSIS



```python
plt.figure(figsize=(12, 4))

# Quantity
plt.subplot(1, 2, 1)
sns.histplot(df['Quantity'], bins=5, kde=True)
plt.title('Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Count')
# TotalPrice
plt.subplot(1, 2, 2)
sns.histplot(df['TotalPrice'], bins=20, kde=True)
plt.title('TotalPrice Distribution')
plt.xlabel('TotalPrice ($)')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python

# Boxplots for outliers
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[['Quantity', 'TotalPrice']])
plt.title('Boxplot of Quantity and TotalPrice')
plt.show()

# Categorical: Product, Category, Location
plt.figure(figsize=(15, 10))

# Product
plt.subplot(3, 1, 1)
sns.countplot(y='Product', data=df, order=df['Product'].value_counts().index)
plt.title('Product Frequency')

# Category
plt.subplot(3, 1, 2)
sns.countplot(y='Category', data=df, order=df['Category'].value_counts().index)
plt.title('Category Frequency')

# Location
plt.subplot(3, 1, 3)
sns.countplot(y='Location', data=df, order=df['Location'].value_counts().index)
plt.title('Location Frequency')

plt.tight_layout()
plt.show()

```

```python
# Date (daily transaction count)
daily_transactions = df.groupby(df['Date'].dt.date).size()
plt.figure(figsize=(10, 5))
daily_transactions.plot(kind='line', title='Daily Transaction Count (April 2025)')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
